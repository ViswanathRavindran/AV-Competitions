```python
Traindf = pd.DataFrame(Traindf[cols])
Testdf = pd.DataFrame(Testdf[cols])
```

```python
# creating X
X = Traindf.iloc[:,1:].as_matrix().astype(np.int)
X
```

```python
y = All_y.as_matrix()
y
```

```python
print ('The shape of XTrain is:',X.shape, 'The shape of y is', y.shape)
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
from sklearn.metrics import r2_score
from sklearn import ensemble
reg1 = ensemble.RandomForestRegressor(verbose=1,random_state=random_state,n_jobs=-1,max_features='sqrt',
                                       n_estimators=1000, min_samples_split=2)
reg1 = reg1.fit(X_train, y_train)
rfpred = pd.Series(reg1.predict(X_test))
```

```python
print ('The R2 score after applying RandomForest is',r2_score(y_test, rfpred) )
```

```python
from sklearn.linear_model import Lasso, ElasticNet
```

```python
reg3 = Lasso(alpha=0.001, fit_intercept=False, normalize=True,max_iter=1000,
             random_state=random_state, positive=True)
reg3 = reg3.fit(X_train, y_train)
LSpred = pd.Series(reg3.predict(X_test))
```

```python
print ('The R2 score after applying Lasso is',r2_score(y_test, LSpred))
```

```python
reg4 = ElasticNet(alpha=0.1, max_iter=10000, l1_ratio=0.2,random_state=random_state, positive=True,
                 normalize=True)
reg4 = reg4.fit(X_train, y_train)
ENpred = pd.Series(reg4.predict(X_test))
```

```python
print ('The R2 score after applying ElasticNet is',r2_score(y_test, ENpred))
```

```python
import patsy
from pyearth import Earth
model = Earth(max_terms=10, max_degree=9, endspan_alpha=0.7, penalty=4.0, 
              verbose=verbose,enable_pruning=True, feature_importance_type=None )
#endspan=, minspan_alpha=, minspan=, thresh=,zero_tol=, smooth=,
```
