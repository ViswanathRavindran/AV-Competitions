```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import sys
pd.set_option('display.max_columns', 99)
pd.set_option('display.max_rows', 999)
verbose = 4
random_state=7
%env JOBLIB_TEMP_FOLDER=/tmp
```

```python
Traindf = pd.read_csv('/prodata/ClassifierTrain3.csv')
All_y = pd.read_csv('/prodata/All_y3.csv')
#filename = '/Users/viswanathravindran/AVDatahack/ADClick/ProData/ClassifierTrain3.csv'
#Traindf = pd.read_csv(filename)
#yfile = '/Users/viswanathravindran/AVDatahack/ADClick/ProData/All_y3.csv'
#All_y = pd.read_csv(yfile)
```

```python
print ('Before One hit encoding', Traindf.shape, All_y.shape)
```

```python
# Convert TrafficType, Browser, OS, Weekday, Dayhour into Numerical
Traindf = pd.get_dummies(Traindf, columns=['TrafficType','DayPeriod','Country','CountryFrd','Carrier','CarrierFrd',
                                           'Device', 'DeviceFrd', 'Browser', 'BrowserFrd', 'AdvCampaign',
                                           'AdvCampaignFrd', 'AndroidOS', 'IPClass'], drop_first=True)
print ('After One hot encoding',Traindf.shape, All_y.shape)
```

```python
Traindf.head()
```

```python
yTrain = All_y['ConversionStatus'].as_matrix().astype(np.int)
del All_y
```

```python
#filename = '/Users/viswanathravindran/AVDatahack/ADClick/ProData/Finaltest.csv'
#Testdf = pd.read_csv(filename)
Testdf = pd.read_csv('/prodata/Finaltest.csv')
Testdf.shape
```

```python
# Convert TrafficType, Browser, OS, Weekday, Dayhour into Numerical
Testdf = pd.get_dummies(Testdf, columns=['TrafficType','DayPeriod','Country','CountryFrd','Carrier','CarrierFrd',
                                           'Device', 'DeviceFrd', 'Browser', 'BrowserFrd', 'AdvCampaign',
                                           'AdvCampaignFrd', 'AndroidOS', 'IPClass'], drop_first=True)
```

```python
print ('Dataframe shapes after encoding the columns', Traindf.shape, Testdf.shape)
```

```python
# Add the missing columns in Test
Testdf['DayPeriod_MonAM1'],Testdf['DayPeriod_MonAM2'],Testdf['DayPeriod_MonAM3'],Testdf['DayPeriod_MonAM4']=0,0,0,0
Testdf['DayPeriod_MonPM1'],Testdf['DayPeriod_MonPM2'],Testdf['DayPeriod_MonPM3'],Testdf['DayPeriod_MonPM4']=0,0,0,0
Testdf['DayPeriod_SunAM1'],Testdf['DayPeriod_SunAM2'],Testdf['DayPeriod_SunAM3'],Testdf['DayPeriod_SunAM4']=0,0,0,0
Testdf['DayPeriod_SunPM1'],Testdf['DayPeriod_SunPM2'],Testdf['DayPeriod_SunPM3'],Testdf['DayPeriod_SunPM4']=0,0,0,0
Testdf['DayPeriod_TueAM1'],Testdf['DayPeriod_TueAM2'],Testdf['DayPeriod_TueAM3'],Testdf['DayPeriod_TueAM4']=0,0,0,0
Testdf['DayPeriod_TuePM1'],Testdf['DayPeriod_TuePM2'],Testdf['DayPeriod_TuePM3'],Testdf['DayPeriod_TuePM4']=0,0,0,0
```

```python
print ('Dataframe shapes after adding the missed column names', Traindf.shape, Testdf.shape)
```

```python
# Import the features 
importances = pd.read_csv('importances175.csv', usecols=range(1,176))
cols = list(importances.columns.values)
print (len(cols))
```

```python
# creating X
NTraindf = Traindf[cols]
XTrain = NTraindf.as_matrix().astype(np.int)
XTrain
```

```python
del Traindf
```

```python
np.bincount(yTrain)
```

```python
print ('The shape of XTrain is:',XTrain.shape, 'The shape of y is', yTrain.shape)
```

```python
# Having the Classifier to learn
from sklearn import ensemble
clf = ensemble.RandomForestClassifier(verbose=verbose,random_state=random_state,n_jobs=-1,max_features='sqrt',
                                       n_estimators=30, min_samples_split=2)
print ('Initializing for training')
start_time = time.time()
clf.fit(XTrain, yTrain)
end_time = time.time()
print ("Training complete in: ", end_time-start_time, "seconds")
```

```python
NTestdf = Testdf[cols]
NTestdf.shape
```

```python
# creating X
XTest1 = NTestdf.iloc[:6200000, :].as_matrix().astype(np.int)
XTest1.shape
```

```python
print ('Predicting it for the 1st Test data')
start_time = time.time()
y_pred1 = (clf.predict_proba(XTest1)[:,1]>0.85).astype(int)
end_time = time.time()
print ('Time taken to fit the First Test set',end_time-start_time, "seconds" )
```

```python
ID1 = pd.DataFrame(Testdf.iloc[:6200000, :1])
ClassSol1 = pd.DataFrame({"ID":ID1.ID, "y":y_pred1})
ClassSol1.to_csv('ClassSol1.csv', index=False)
del ClassSol1, XTest1, ID1
```
